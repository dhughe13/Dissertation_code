```python
Jn_025.shape
```

```python
def FC(Jn_num):
    Jn_FC = np.corrcoef(Jn_num.T)
    return Jn_FC

def Pearson(Jn_FC,emp_FC):
    emp_FCf = np.ndarray.flatten(emp_FC)
    Jn_FCf = np.ndarray.flatten(Jn_FC)
    R = np.corrcoef(Jn_FCf,emp_FCf)
    Pearson = R[0,1]
    return Pearson
```

```python
Jn_val = [0.025, 0.075, 0.125, 0.175, 0.225, 0.275, 0.325, 0.375, 0.425, 0.475, 0.5]

Jn025_FC = FC(Jn_025)
Jn025_P = Pearson(Jn025_FC,emp_FC)

Jn075_FC = FC(Jn_075)
Jn075_P = Pearson(Jn075_FC,emp_FC)

Jn125_FC = FC(Jn_125)
Jn125_P = Pearson(Jn125_FC,emp_FC)

Jn175_FC = FC(Jn_175)
Jn175_P = Pearson(Jn175_FC,emp_FC)

Jn225_FC = FC(Jn_225)
Jn225_P = Pearson(Jn225_FC,emp_FC)

Jn275_FC = FC(Jn_275)
Jn275_P = Pearson(Jn275_FC,emp_FC)

Jn325_FC = FC(Jn_325)
Jn325_P = Pearson(Jn325_FC,emp_FC)

Jn375_FC = FC(Jn_375)
Jn375_P = Pearson(Jn375_FC,emp_FC)

Jn425_FC = FC(Jn_425)
Jn425_P = Pearson(Jn425_FC,emp_FC)

Jn475_FC = FC(Jn_475)
Jn475_P = Pearson(Jn475_FC,emp_FC)

Jn5_FC = FC(Jn_5)
Jn5_P = Pearson(Jn5_FC,emp_FC)
```

```python
Jn_P = [Jn025_P, Jn075_P, Jn125_P, Jn175_P, Jn225_P, Jn275_P, Jn325_P, Jn375_P, Jn425_P, 
        Jn475_P, Jn5_P]
np.save('Jn_P-10069.npy',Jn_P)
```
