```python
def FC(wp_num):
    wp_FC = np.corrcoef(wp_num.T)
    return wp_FC

def Pearson(wp_FC,emp_FC):
    emp_FCf = np.ndarray.flatten(emp_FC)
    wp_FCf = np.ndarray.flatten(wp_FC)
    R = np.corrcoef(wp_FCf,emp_FCf)
    Pearson = R[0,1]
    return Pearson
```
