```python
def FC(Ji_num):
    Ji_FC = np.corrcoef(Ji_num.T)
    return Ji_FC

def Pearson(Ji_FC,emp_FC):
    emp_FCf = np.ndarray.flatten(emp_FC)
    Ji_FCf = np.ndarray.flatten(Ji_FC)
    R = np.corrcoef(Ji_FCf,emp_FCf)
    Pearson = R[0,1]
    return Pearson
    
```
